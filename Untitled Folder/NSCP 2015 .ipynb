## Period of Vibration

```python
import math

height = 12.300*3.280839895 ##meters 
d = 3.008*3.280839895 ##meters
w = 599*0.22480894387096 ##kN 
t = (12/1000)*3.280839895 ##mter
oc = 1
def period(oc,d,w,t):
    A = (w*d)/t
    constant = 7.6e-6
    value = constant*pow(oc/d,2)*math.sqrt(A)
    return value

print(period(oc,d,w,t))
```
