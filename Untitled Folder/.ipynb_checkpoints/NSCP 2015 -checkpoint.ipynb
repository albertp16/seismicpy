## Period of Vibration

```python
import math

height = 12.300 ##meters
d = 3.008 ##meters
w = 599 ##kN 
t = 12/1000 ##mter
oc = 1
def period(oc,d,w,t):
    A = (w*d)/t
    constant = 7.6e-6
#     constant = 2.04/10e5
    value = constant*pow(oc/d,2)*math.sqrt(A)
    return value

print(period(oc,d,w,t))
```
