### Point Resistance (Qp) 
reference: Vesic (1977)
#### Earth Pressure Coefficient at rest
<center> $ K_{o} = 1 - \text{sin}\left( \phi^{'} \right) $ </center>

#### Mean Effective Normal Ground stress at the Level of the Pile Point
<center> $ \bar{ \sigma_{o} } = 1 - \left( \frac{1 + 2K_{o}}{3} \right) $ </center>

#### Rigidity Index for the Soil 
<center> $ I_{rr} = \frac{E_{s}}{2 \left( 1 + \mu _{s} \right) q^{'} \text{tan} \phi^{'} } $ </center>

#### Reduced Rigidity Index for the Soil 
<center> $ I_{rr} = \frac{I_{r}}{1 + I_{r} \Delta } $ </center>

```python
def ko(af):
    angle = math.radians(af)
    value = 1 - math.sin(angle)
    return value

def mus(af):
    value = 0.1 + 0.3*((af - 25)/20)
    return value

def sigmaO(q_prime,af):
    value = ((1 + 2*(1 - math.sin(math.radians(af))))/3)*q_prime
    return value

def delta(af,q_prime):
    value = 0.005*(1 - ((af - 25)/20))*(q_prime/100)
    return value

def Ir(Es,mu_s,q_prime,af):
    value = Es/(2*(1+mu_s)*q_prime*math.tan(math.radians(af)))
    return value

def Irr(ir,delta):
    value = ir/(1 + (ir*delta))
    return value

q_prime = 18*5

Es = 100*200
af = 35 
mu_s = mus(af)
sigma_o = sigmaO(q_prime,af)
delta_value = delta(af,q_prime)
ir = Ir(Es,mu_s,q_prime,af)
irr = Irr(ir,delta_value)

print('Sigma_o = ' + str(sigma_o))
print('Delta = ' + str(delta_value))
print('Ir = ' + str(ir))
print('Irr = ' + str(irr))
nq = 83.78
ap = (math.pi/4)*math.pow(0.3,2)
qp = ap*nq*sigma_o #Point Resistance
```

### Skin Friction

```python

sbc = [50,50,50,50,50,58.76,209.26,350.65,350.65,350.65]
ks = []

for i in sbc: 
    ks_index = i*120
    ks.append(ks_index)
# print(ks)

sk_arr = []
perimeter = math.pi*0.40 #meters
for i in ks:
    index = i*perimeter
    sk_arr.append(index)
# print(sk_arr)
sk_df = pd.DataFrame({
      'height (m)': ht,
      'skin friction (kN/m)': sk_arr,  
      })
sk_df
```

```python
hsgr = []
for i in ks:
    hsgr_index = i*0.40*0.5
    hsgr.append(hsgr_index)

sk_df = pd.DataFrame({
      'height (m)': ht,
      'Lateral spring (kN/m)': hsgr,  
      })
sk_df
```
