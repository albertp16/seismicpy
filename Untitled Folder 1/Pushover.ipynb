```python
import math

def target_displacement(c0,c1,c2,c3,te):
    '''
    Calculates the target displacement
    :param float c0: Modification Factor related spectral displacement
    :param float c1: Modification Factor related maximum inelastic displacment
	:param float c2: Modification Factor represent effect of pinched hysteretic shape, stiffness degradation and strength deterioration on maximum displacement response. 
    :param float c3: Modification Factor increased displacement dynamic P-Delta Effect
    :param float te: Effective fundamental period
    '''
    GRAVITY = 9.81
    delta_t = c0*c1*c2*c3*((math.pow(te,2))/(4*math.pow(math.pi,2)))*GRAVITY
    return delta_t


target_displacement(1.2,2,3,4,5)
```
