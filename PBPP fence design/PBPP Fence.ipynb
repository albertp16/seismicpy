## Base Shear

```python
I = 1.0 
nv = 1.00 #From Table 208-5
ca = 0.44 #From Table 208-7
W = 0.867 ## kN
Z = 0.4
R = 2.2

V_208025 = 0.56*ca*I*W

V_208026 = ((1.6*Z*nv*I)/R)*W
```
