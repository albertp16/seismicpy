```python
import apecseismicpy

distance, source_type, soil_type, zone = 2, "A", "sd", 4
site_response = apecseismicpy.site_coefficients(distance, source_type, soil_type,zone)
site_data = site_response.calculate()

na = site_data['na']
nv = site_data['nv']
ca = site_data['ca']
cv = site_data['cv']

print(na)
print(nv)
print(ca)
print(cv)

plot = apecseismicpy.ResponseSpectrum(ca,cv)
plot.plot()
```
