```python
import apecseismicpy as sd
# import apecseismicpy.aci350.hydrodynamic as sd
import math

tslab = 150 #mm
tw = .500 #mm
fc = 35 #Mpa
ec = 4700*math.sqrt(fc) #Mpa 
gamma_l = 10.10 #Kn/m3, unit weight of liquid
gamma_c = 24 #Kn/m3, unit weight of concrete

l = 14.5 #m, tank dimension perpendicular 
b = 7.2 #m, tank dimension perpendicular 
hl = 7.15 #m, liquid content depth
liquid_weight = l*b*hl*gamma_l
print(liquid_weight)

print("Weight of Accelarting Liquids Section")
tank = sd.effective_liquid_weights(l,hl,liquid_weight)
wi = tank['impulsive']['value']
print(tank)
print("Weight of Accelarting Liquids Section")
contriods = sd.calculate_heights_of_centers_of_gravity(l,hl)

hi = float(contriods["EPB"]["hi"])
print(contriods)

hw = 8 #m Wall Height

properties = sd.DynamicProperties(l,hw,tw,wi,liquid_weight,hl,hi,ec,gamma_c,gamma_l)
mw_i =  properties.compute_mw()
mi_i =  properties.compute_mi()
h_i =  properties.compute_h()
k_i =  properties.compute_k()
mt_i =  properties.compute_mt()
ti_i =  properties.compute_ti()
tc_i =  properties.compute_tc()
print(mw_i)
print(mi_i)
print(h_i)
print(k_i)
print(mt_i)
print(ti_i)
print(tc_i)
```
